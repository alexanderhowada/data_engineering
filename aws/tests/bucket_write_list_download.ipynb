```python
import boto3
```

```python
s3 = boto3.resource('s3')
bucket = s3.Bucket('ahow-delta-lake')

for obj in bucket.objects.all():
    print(obj.key)
```

```python
new_object = 'new_obj.txt'
bucket.put_object(Body='content', Key=new_object)
bucket.download_file(new_object, 'asdf.txt')
```
